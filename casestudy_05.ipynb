There are missing values in age, Cabin and Embarked. Out of this, Cabin details is missing from majority of the passengers.

# 4. Fill in all the missing values present in all the columns in the dataset

```python
data['Age']=data['Age'].fillna(data['Age'].median())
data
```

```python
data['Embarked']= data['Embarked'].fillna(data['Embarked'].mode()[0])
data
```

When it comes to column containing cabin number, majority of data is missing. Since this is a detail which cannot be filled using mean, median or mode (as passengers stay in individual cabins) and as more than 60% of data is missing, it is better to drop the column (It is always better to discuss with the stakeholders on how a particular column affects data analysis before deciding to drop the column.)

```python
data.drop('Cabin', inplace=True, axis=1)
data
```

# 5. Check and handle outliers in at least 3 columns in the dataset

# Handling outliers for column 'Age'

```python
plt.boxplot(data["Age"])
plt.title("Boxplot of Age",fontsize=15)
```

```python
q1=np.percentile(data["Age"],25)
q3=np.percentile(data["Age"],75)
IQR=q3-q1
print("Q1 =",q1)
print("Q3 =",q3)
print("IQR =",IQR)
low_lim=q1-1.5*IQR
high_lim=q3+1.5*IQR
outlier1=data.loc[(data["Age"]>high_lim)]
outlier2=data.loc[(data["Age"]<low_lim)]
print("The outliers are",outlier1,outlier2)
outlier1_index=outlier1.index
outlier2_index=outlier2.index
```

```python
data1=data.drop(outlier1_index)
data2=data1.drop(outlier2_index)
data2
```

# Handling outliers of column 'Fare'

```python
plt.boxplot(data2["Fare"])
plt.title("Boxplot of Fare",fontsize=15)
```

```python
q_1=np.percentile(data2["Fare"],25)
q_3=np.percentile(data2["Fare"],75)
IQR1=q_3-q_1
print("Q1 =",q_1)
print("Q3 =",q_3)
print("IQR =",IQR1)
lower_lim=q_1-1.5*IQR1
higher_lim=q_3+1.5*IQR1
outlier_1=data2.loc[(data2["Fare"]>higher_lim)]
outlier_2=data2.loc[(data2["Fare"]<lower_lim)]
print("The outliers are",outlier_1,outlier_2)
outlier_1_index=outlier_1.index
outlier_2_index=outlier_2.index
```

```python
data3=data2.drop(outlier_1_index)
data4=data3.drop(outlier_2_index)
data4
```
